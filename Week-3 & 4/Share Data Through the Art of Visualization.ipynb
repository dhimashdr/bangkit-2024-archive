# Share Data Through the Art of Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/dhimashdr/akreditasi-univ-dummy/main/data_pmm_dummy_fix_final.csv")
```

```python
data_amazon = pd.read_excel("https://github.com/dhimashdr/bangkit-2024-archive/raw/main/Week-3%20&%204/Amazon%202_Raw.xlsx")
```

```python
by_keketatan = data.groupby(by="Keketatan").size().reset_index(name='Jumlah')
by_keketatan
```

## Type of Visualizations

### Bar Plot

```python
plt.figure(figsize=(10, 4))
 
sns.barplot(
    y="Jumlah", 
    x="Keketatan",
    data=by_keketatan
)
plt.title("Sebaran Keketatan Seleksi Perguruan Tinggi", loc="center", fontsize=15)
plt.xlabel("Keketatan")
plt.ylabel("Jumlah")
plt.show()
```

```python
sales_by_year = data_amazon.resample(rule='Y', on='Order Date').Sales.sum().reset_index(name="Sales")
sales_by_year
```

### Line Plot

```python
plt.figure(figsize=(10,5))

sns.lineplot(
    data = sales_by_year,
    x = "Order Date",
    y = "Sales"
)

plt.title("Annual Sales", loc="center", fontsize=14)
plt.xlabel("Order Date")
plt.ylabel("Sales")
plt.show()
```

### Pie Chart

```python
keketatan_key = list(by_keketatan.Keketatan)
keketatan_value = list(by_keketatan.Jumlah)
```

```python
plt.figure(figsize=(8,8))

plt.title("Pie Chart Keketatan", fontsize=15)
plt.pie(keketatan_value, labels = keketatan_key, explode = (0, 0.1, 0), autopct='%1.1f%%')
plt.show()
```

Usually, the y-axis started from 0

## Dynamic Visualization

Dynamic visualization is a visualization that the graph, plot, etc. are interactive.

**Types of visualization**

- Line chart ~ used to track changes over a period time
- Column chart ~ spt bar chart, tpi vertically, use size to contrast and compare two or more values
- Heatmap ~ to compare categories by using color range
- Pie chart ~ proportions corresp to quantty (percentage)
- Scatterplot ~ show relationship between different variables
- Distance graph ~ display the spread of vary outcomes in a dataset

## Decision to choose type of visualization

If data has only one numeric variable
- Histo
- Density plot

Multiple dataset
- Line chart
- Pie chart

Measuring changes over time

Need to show the relationship between the data

## Principles of Design

- Balance : balancing the key visual elements, like color and shape are distributed evenly.
- Emphasis : our data viz should have focal point, in order to make reader focus on the visualization.
- Movement : refer to the path viewer read line. (in chart is from left to right)
- Pattern : use a consistent shapes and colors.
- Repetition : repetitive of using shape, color, etc in order to make visualization effectively.
- Proportion : how the data viz proportion is comfort to see.
- Rhythm : same as movement
- Variety : it's kind of the contra of repetitive, sometimes it better to make various type of visualization.
- Unity : it should well organized.

## Design Thinking

- Empathize : Think about the emotions and needs of the target audience.
- Define : Understand the audiences needs, problems, and insights.
- Ideate : Use your findings from the previous phases to begin to create data visualizations.
- Prototype : Start putting it all together. In this case, you can put your findings into  a presentation or dashboard.
- Test : Check that your prototype is effective. In this case, you can show your visualizations to team members before the presentation.

*user-centric way

Don't forget about **5 sec rule**
\
If you viz is easy to understand by your audience within 5 seconds, then your viz is good.

**Tips :**
\
<font color='red'>**Red</font>-<font color='lime'>green</font>** color blindness is the most common and occurs when red and green look like the same color. You can avoid placing green on red or red on green in your visualizations. 
\
<font color='blue'>**Blue</font>-<font color='yellow'>yellow</font>** color blindness is less common and occurs when it is difficult to tell the difference between blue and green, or yellow and red. You can also avoid using these colors on top of or next to each other.
